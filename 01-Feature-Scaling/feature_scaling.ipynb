# Feature Scaling 

- **Definition:** Adjusting numeric features so they share a common scale.
- **Purpose:** Prevents features with larger ranges from dominating model training.
- **When to use:** For models that rely on distance or gradient-based optimization (e.g., k-NN, SVM, logistic regression, neural networks).
- **Common methods:** Standardization (z-score), Min–Max normalization.
- **Tip:** Fit the scaler on the training set and apply it to validation/test sets to avoid data leakage.

# Why Feature Scaling Matters

**Problem:** The `salary` column has values much larger than `mba_p` and `etest_p`. When features have different scales, many models (especially distance-based or gradient-based algorithms) treat larger-valued features as more important.

- **Effect:** Models can become dominated by `salary`, producing biased weights and poor generalization.
- **Goal:** Give each feature equal importance by putting them on a comparable scale.

**Common Solutions:**
- **Standardization** — subtract mean and divide by standard deviation (`StandardScaler`).
- **Min–Max Scaling** — rescale to a specific range, usually [0, 1] (`MinMaxScaler`).

**Best Practice:** Fit the scaler on the training set and apply it to the test set to avoid data leakage.

# **Normalization, MIN MAX Scaling**

- Normalization is a scalin technique in which values are shifted and rescaled.
- so that they end up rangin between 0 and 1. It is also known as Min Max Scalingg

**Formula (rescale to [0, 1]):**

$$
x' = \frac{x - x_{\min}}{x_{\max} - x_{\min}}
$$

**General form (range [a, b]):**

$$
x' = a + \frac{(x - x_{\min})(b - a)}{x_{\max} - x_{\min}}
$$

*If $x_{\max} = x_{\min}$ (constant feature), set $x' = a$ (or $0$ for $[0,1]$) to avoid division by zero.*

# When to apply feature scaling

- **K-Nearest Neighbors (KNN):** KNN relies on distance metrics; scaling significantly affects performance.
- **Support Vector Machines (SVM):** SVM uses distances between points; feature scaling is essential for correct margins.
- **K-Means Clustering:** Clustering depends on distances; unscaled features bias cluster assignments.
- **Principal Component Analysis (PCA):** PCA is sensitive to feature variances; scale features before computing components.
- **Gradient Descent–based algorithms:** Models like Linear/Logistic Regression and Neural Networks converge faster and more reliably with scaled inputs.
- **Gaussian Mixture Models (GMM):** Like K-Means, GMMs use distance metrics; scaling helps model the underlying distribution correctly.

**Quick rule of thumb:** If the algorithm uses distances or gradient-based optimization, apply feature scaling.

# Feature Scaling — Standardization (Z-score)

**What it is:**
- Standardization rescales features so they have mean 0 and standard deviation 1.
- Also called *z-score normalization*.

**Formula (per feature):**

$$
z = \frac{x - \mu}{\sigma}
$$

where:
- $x$ is the original feature value,
- $\mu$ is the mean of the feature (usually computed on the training set),
- $\sigma$ is the standard deviation of the feature (usually computed on the training set).

To invert the transform (recover original values):

$$
x = z\sigma + \mu
$$

**Why use standardization:**
- Centers features around zero and gives them unit variance, which helps many algorithms behave better.
- Improves convergence speed for gradient-based optimizers (e.g., neural networks, logistic regression).
- Makes regularization penalties (L1, L2) act consistently across features.
- Necessary for algorithms that assume standardized input (e.g., PCA, some implementations of SVM).

**When to prefer standardization vs Min–Max:**
- Use **standardization** when your data is Gaussian-like or your model assumes zero-centered inputs (linear models, logistic regression, neural networks, PCA, SVM).
- Use **Min–Max scaling** when you need a bounded range (e.g., [0,1]) or when features are already roughly uniformly distributed.
- If your data contains many outliers, consider **RobustScaler** (uses median and IQR) instead of StandardScaler.

**Best practices:**
- Always *fit* the scaler on the training set only, then *transform* the validation/test sets to avoid data leakage:

- After standardization, inspect means and variances to confirm transformation (training mean ≈ 0, training std ≈ 1).

**Summary:**
- Standardization (z-score) centers features and sets unit variance. Use it for models sensitive to feature centering/scaling and for better optimizer behavior; prefer robust methods if outliers dominate.
